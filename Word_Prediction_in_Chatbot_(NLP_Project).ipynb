```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from collections import deque

# Initialize GPT-2 with caching
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.eval()

# Context window (last 5 words)
context_window = deque(maxlen=5)

def get_predictions(full_text, num_suggestions=3):
    """Predict next word using sentence-level context"""
    inputs = tokenizer.encode(full_text, return_tensors='pt')

    with torch.no_grad():
        outputs = model.generate(
            inputs,
            max_length=inputs.shape[1] + 1,
            num_return_sequences=num_suggestions,
            do_sample=True,
            top_k=50,
            temperature=0.7
        )

    return list(set(
        tokenizer.decode(output[-1], skip_special_tokens=True)
        for output in outputs
    ))

# Simulate typing with context
current_sentence = ''
print("Type a sentence (space to predict, Enter to exit):")
while True:
    user_input = input(f"Current: {current_sentence}\nNext char: ")

    # Check if the input is empty (user pressed Enter)
    if not user_input:
        break

    # Get the first character of the input
    user_input = user_input[0]

    if user_input == '\n':
        break

    if user_input == ' ':  # Space triggers prediction
        suggestions = get_predictions(current_sentence.strip())
        print(f"Suggestions: {suggestions}")
        current_sentence += ' ' + suggestions[0] if suggestions else ' '
    else:
        current_sentence += user_input
```
